## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
btc_response_data = requests.get(btc_url)
btc_json_data = btc_response_data.json()

# Display the data in JSON format
print(json.dumps(btc_json_data, indent=4))
```

```python
# Parse the JSON data and fetch current maket price (CMP) of Bitcoin
btc_cmp = btc_json_data['data']['1']['quotes']['USD']['price']
```

```python
# Fetch current ETH price
eth_response_data = requests.get(eth_url)
eth_json_data = eth_response_data.json()

# Display the data in JSON format
print(json.dumps(eth_json_data, indent=4))
```

```python
# Parse the JSON data and fetch current maket price (CMP) of Ethereum
eth_cmp = eth_json_data['data']['1027']['quotes']['USD']['price']
```

```python
# Compute current value of my crpto = amount * CMP
my_btc_value = my_btc * btc_cmp
my_eth_value = my_eth * eth_cmp

# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
# Format current date as ISO format
current_date = pd.Timestamp("2021-05-25", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# Note that both 'start' and 'end' are set to the current_date
df_ticker = api.get_barset(
    tickers,
    timeframe,
    start=current_date,
    end=current_date,
    limit=1000
).df

# Preview DataFrame
df_ticker
```

```python
# Pick AGG and SPY close prices
# df_ticker["AGG"]["close"] will slice AGG close price as a dataframe with datetime index and close price as 
# a series or array, even though it's a single value. 
# .values will get rid of datetime index from the selected slice and int() will convert array of single close 
# price to a simple int => converting [114.74] to 114.74

agg_close_price = int(df_ticker["AGG"]["close"].values)
spy_close_price = int(df_ticker["SPY"]["close"].values)

print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares = amount * close price
my_spy_value = my_spy * spy_close_price
my_agg_value = my_agg * agg_close_price

# Print current value of share
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income = 12000

# Create savings DataFrame

# Compute crypto value and round to 2 decimal places
crypto_value = my_btc_value + my_eth_value
crypto_value = round(crypto_value,2)

# Compute shares value and round to 2 decimal places
shares_value = my_spy_value + my_agg_value
shares_value = round(shares_value,2)

# Construct savings dataframe

# Packing the data into a dictionary
data = {
    'asset_class': ['crypto','shares'],
    'amount': [crypto_value,shares_value]
}

# Collapsing the dictionary into a dataframe 
df_savings = pd.DataFrame(data)

# Set 'asset_class' as index of the dataframe
df_savings.set_index('asset_class', inplace=True) 

# Display savings DataFrame
df_savings
```

```python
# Plot savings pie chart
df_savings.plot.pie(y="amount", title="Composition of Personal Savings")
```

```python
# Set ideal emergency fund = 3 months of monthly income
emergency_fund = monthly_income * 3

# Calculate total amount of savings in crypto and shares
total_savings = crypto_value + shares_value

# Validate saving health
if total_savings > emergency_fund:
    print(f"Congratulations! You have enough money in your emergency fund.")
elif total_savings == emergency_fund:
    print(f"Congratulations! Your savings is just enough to meet your emergency fund.")
else:
    print(f"Your savings is ${emergency_fund - total_savings} away from emergency fund.")
    
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date = pd.Timestamp('2016-05-25', tz='America/New_York').isoformat()
end_date = pd.Timestamp('2021-05-25', tz='America/New_York').isoformat()
```

```python
# Get 5 years' worth of historical data for SPY and AGG
df_stock_data = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df


# Display sample data
df_stock_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns

# Set number of simulations
num_sims = 500

# Set portfolio weights - 40% AGG and 60% SPY. 
# tickers = ["AGG", "SPY"]
portfolio_weights = [0.40, 0.60]

# Configure a Monte Carlo simulation to forecast 30 years daily returns
MC_30years = MCSimulation(
    portfolio_data = df_stock_data,
    weights = portfolio_weights,
    num_simulation = num_sims,
    num_trading_days = 252*30
)
```

```python
# Printing the simulation input data
MC_30years.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30years.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_30years.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_30years.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
summary_stat = MC_30years.summarize_cumulative_return()

# Print summary statistics
print(summary_stat)
```

### Calculate the expected portfolio return at the 95% lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(summary_stat[8]*initial_investment,2)
ci_upper = round(summary_stat[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower = round(summary_stat[8]*initial_investment,2)
ci_upper = round(summary_stat[9]*initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
